```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('dataset/train.csv')
```

```python
test = pd.read_csv('dataset/test.csv').drop(['ID_code'],axis=1)
```

```python
trainX_data = data.drop(['ID_code','target'],axis=1)
```

```python
trainY_data = data['target']
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
trainX, valX, trainY, valY = train_test_split(trainX_data,trainY_data,test_size=0.3)
```

```python
train = lgb.Dataset(trainX,label=trainY)
val = lgb.Dataset(valX,label=valY)
```

```python
params= {

 'n_estimators': [1000],
  
  'num_leaves': [4,5,6],
  'min_data_in_leaf': [68,70,72],
  'feature_fraction': [0.1,0.2,0.3],
  'bagging_fraction': [0.1,0.2,0.3]
    
    
  

}
```

```python
params = {
    'num_leaves': [6,8,10,15],
    'max_depth' : [-1],
    'min_data_in_leaf': [70,80,85,90],
    'learning_rate': [0.01],
    'feature_fraction': [0.1],
    'bagging_fraction': [0.2],
    'bagging_freq': [3,5,7,9]

}
```

```python
gbm = lgb.LGBMClassifier(objective='binary',metric='auc')
```

```python
gsearch = GridSearchCV(gbm, param_grid=params, scoring='accuracy',cv=3)
```

```python
gsearch.fit(trainX,trainY)
print('best score:' ,gsearch.best_score_)
```

```python
gsearch.best_estimator_.get_params()
```
