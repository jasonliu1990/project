```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from itertools import chain
```

## 食譜爬蟲
1. 爬取網站內所有分類的頁面
2. 爬取各頁面下的所有page
3. 取得所有食譜的組成

```python
url = 'https://icook.tw/categories'
header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.96 Safari/537.36'}

html_doc = requests.get(url,headers = header).text
html_doc = BeautifulSoup(html_doc, 'html.parser')
```

```python
url_list = html_doc.select('.list-group-item~ .list-group-item+ .list-group-item .list-group-item a , .featured-cards~ .inner-block .list-title')
url_list = list(map(lambda x: x.get('href'), url_list))
url_list = list(map(lambda x: 'https://icook.tw' + x, url_list))
print(len(url_list))
```

```python
nonext_title = '愛料理上找不到您要的頁面喔～'
recipe_title = []
compose = []
for i in range(len(url_list)):
    
    page_no = 1
        
    while (True):
        url = url_list[i]
        url = url + '?page=' + str(page_no)
        
        doc = requests.get(url,headers = header).text
        doc = BeautifulSoup(doc, 'html.parser')
        title = doc.select('head title')[0].get_text()
              
        if title == nonext_title:
            
            break
            
        else :
            
            tmp1 = list(map(lambda x: x.get_text().strip(), doc.select('.browse-recipe-name')))
            tmp2 = list(map(lambda x: x.get_text().strip(), doc.select('.browse-recipe-content-ingredient')))
            recipe_title.append(tmp1)
            compose.append(tmp2)
            page_no += 1
    
    print(i)
```

```python
compose_list = list(chain(*compose))
compose_list = list(set(compose_list))
compose_list = list(map(lambda x: x.strip('食材：'), compose_list))
print(len(compose_list))
```
